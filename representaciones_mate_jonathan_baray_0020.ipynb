Problema 2: Las edades de Pedro y AnaPaso 1: Planteamiento y resoluciónIdentificar las incógnitas:Edad de Ana: la llamaremos $a$.Edad de Pedro: como es el doble que la de Ana, la representamos como $2a$.Establecer la ecuación: La suma de ambas edades es igual a 36.$$a + 2a = 36$$Simplificar y despejar:Sumamos términos semejantes: $3a = 36$.Para dejar sola a la $a$, el 3 que está multiplicando pasa dividiendo:$$a = \frac{36}{3}$$Calculamos el valor: $a = 12$.Encontrar la segunda edad: Si Ana tiene 12 años, Pedro tiene $2 \times 12 = 24$.Resultado: Ana tiene 12 años y Pedro tiene 24 años.


```python
# Problema 2: Calcular edades de Pedro y Ana
total_suma = 36
suma_total=int(input("ingrese la suma total: "))
# Sabemos que 3 veces la edad de Ana es igual al total (a + 2a = 3a)
edad_ana = total_suma / 3
# La edad de Pedro es el doble de la de Ana
edad_pedro = edad_ana * 2

print(f"La edad de Ana es: {int(edad_ana)} años")
print(f"La edad de Pedro es: {int(edad_pedro)} años")
```
